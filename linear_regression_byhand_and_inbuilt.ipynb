```python
#Linear regression by hand
import numpy as np
import matplotlib.pyplot as plt
import time

# Start time
start = time.time()

# Least Square Method - 1st Sum
x = np.array([95, 85, 80, 70, 60])
y = np.array([85, 95, 70, 65, 70])
n = len(x)

xm = x.mean()  # mean of x
ym = y.mean()  # mean of y

xd = x - xm     # x deviations
yd = y - ym     # y deviations

xy = xd * yd
xysum = xy.sum()

xsq = xd**2
xsq_sum = xsq.sum()

ysq = yd**2
ysq_sum = ysq.sum()

b1 = xysum / xsq_sum  # slope
b0 = ym - b1 * xm     # intercept

print("slope b1 =", b1)
print("intersect b0 =", b0)

# Coefficient of correlation
r_den = xsq_sum * ysq_sum
rxy = xysum / np.sqrt(r_den)
print("coefficient of correlation =", rxy)

# Coefficient of determination
Rsq = rxy**2
print("coefficient of determination =", Rsq)

# Cost Function
yg = b1 * x + b0
temp = np.sum(np.square(y - yg))
cost = (1 / n) * temp
print("cost function =", cost)

# Plot
plt.scatter(x, y)
plt.plot(x, yg, 'D-r')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Least Square Method")
plt.show()

# End time
end = time.time()

# Print execution time
print(f"Execution Time: {end - start:.6f} seconds")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.stats import linregress
from sklearn.linear_model import LinearRegression

x = np.array([95, 85, 80, 70, 60])
y = np.array([85, 95, 70, 65, 70])

# Convert x to 2D for sklearn
x_reshaped = x.reshape(-1, 1)
```

```python
# Set up plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='black', label='Data Points')
# -----------------------------
# Method 1: NumPy polyfit
# -----------------------------
start = time.time()
b1_np, b0_np = np.polyfit(x, y, 1)
end = time.time()
yg_np = b1_np * x + b0_np
plt.plot(x, yg_np, 'r--', label=f'Numpy polyfit\nTime: {end-start:.6f}s')
plt.title("Linear Regression Methods Comparison")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
print("NumPy polyfit:")
print("  Slope:", b1_np)
print("  Intercept:", b0_np)
print("  Execution Time:", end - start, "s\n")
```

```python
# Set up plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='black', label='Data Points')
# -----------------------------
# Method 2: SciPy linregress
# -----------------------------
start = time.time()
result = linregress(x, y)
end = time.time()
yg_scipy = result.slope * x + result.intercept
plt.plot(x, yg_scipy, 'g-.', label=f'SciPy linregress\nTime: {end-start:.6f}s')
plt.title("Linear Regression Methods Comparison")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
print("SciPy linregress:")
print("  Slope:", result.slope)
print("  Intercept:", result.intercept)
print("  r-value:", result.rvalue)
print("  R²:", result.rvalue**2)
print("  p-value:", result.pvalue)
print("  Execution Time:", end - start, "s\n")
```

```python
# Set up plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='black', label='Data Points')
# -----------------------------
# Method 3: sklearn LinearRegression
# -----------------------------
start = time.time()
model = LinearRegression().fit(x_reshaped, y)
end = time.time()
yg_sklearn = model.predict(x_reshaped)
plt.plot(x, yg_sklearn, 'b-', label=f'Sklearn LinearRegression\nTime: {end-start:.6f}s')
plt.title("Linear Regression Methods Comparison")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
print("Sklearn LinearRegression:")
print("  Slope:", model.coef_[0])
print("  Intercept:", model.intercept_)
print("  R² Score:", model.score(x_reshaped, y))
print("  Execution Time:", end - start, "s\n")
```
