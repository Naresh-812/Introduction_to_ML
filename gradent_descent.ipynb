```python
#gradent descent method
#Without inbuilt librarys
%%time
import numpy as np
import matplotlib.pyplot as plt
x=np.array([95,85,80,70,60])
y=np.array([85,95,70,65,70])
n=np.size(x)
m_in=0
bcur=0

iterations=10000000
learning_rate=0.0001
for i in range(iterations):
  ypred=m_in*x+bcur
  temp=np.sum(np.square(y-ypred))
  cost=(1/n)*temp
  md=-(2/n)*sum((y-ypred)*x)
  bd=-(2/n)*sum(y-ypred)
  m_in=m_in-(learning_rate*md)
  bcur=bcur-(learning_rate*bd)
  print("m {},b {}, cost {},iterations {}".format(m_in,bcur,cost,i))
```

```python
# Gradient Descent using scikit-learn LinearRegression (which internally uses optimized solvers)
#SGDRegressor
from sklearn.linear_model import SGDRegressor
import numpy as np
import matplotlib.pyplot as plt
import time

# Data
x = np.array([95,85,80,70,60]).reshape(-1, 1)  # sklearn expects 2D input for features
y = np.array([85,95,70,65,70])

# Model initialization with gradient descent optimizer
model = SGDRegressor(max_iter=100000, tol=1e-6, learning_rate='constant', eta0=0.0001)

# Timing the fit process
start_time = time.time()
model.fit(x, y)
end_time = time.time()

# Extract slope and intercept
b1 = model.coef_[0]
b0 = model.intercept_[0]

# Predictions
y_pred = model.predict(x)

# Calculate cost (MSE)
cost = np.mean((y - y_pred)**2)

print(f"Slope (b1): {b1}")
print(f"Intercept (b0): {b0}")
print(f"Cost (MSE): {cost}")
print(f"Execution time: {end_time - start_time:.5f} seconds")

# Plotting
plt.scatter(x, y, label='Actual data')
plt.plot(x, y_pred, color='red', label='Fitted line (SGD)')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Gradient Descent using scikit-learn SGDRegressor')
plt.legend()
plt.show()

```
