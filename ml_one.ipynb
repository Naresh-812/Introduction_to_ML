```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Start time
start = time.time()

# Least Square Method - 1st Sum
x = np.array([95, 85, 80, 70, 60])
y = np.array([85, 95, 70, 65, 70])
n = len(x)

xm = x.mean()  # mean of x
ym = y.mean()  # mean of y

xd = x - xm     # x deviations
yd = y - ym     # y deviations

xy = xd * yd
xysum = xy.sum()

xsq = xd**2
xsq_sum = xsq.sum()

ysq = yd**2
ysq_sum = ysq.sum()

b1 = xysum / xsq_sum  # slope
b0 = ym - b1 * xm     # intercept

print("slope b1 =", b1)
print("intersect b0 =", b0)

# Coefficient of correlation
r_den = xsq_sum * ysq_sum
rxy = xysum / np.sqrt(r_den)
print("coefficient of correlation =", rxy)

# Coefficient of determination
Rsq = rxy**2
print("coefficient of determination =", Rsq)

# Cost Function
yg = b1 * x + b0
temp = np.sum(np.square(y - yg))
cost = (1 / n) * temp
print("cost function =", cost)

# Plot
plt.scatter(x, y)
plt.plot(x, yg, 'D-r')
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Least Square Method")
plt.show()

# End time
end = time.time()

# Print execution time
print(f"Execution Time: {end - start:.6f} seconds")
```

```python
#gradent descent method
#1 st problem
%%time
import numpy as np
import matplotlib.pyplot as plt
x=np.array([95,85,80,70,60])
y=np.array([85,95,70,65,70])
n=np.size(x)
m_in=0
bcur=0

iterations=10000000
learning_rate=0.0001
for i in range(iterations):
  ypred=m_in*x+bcur
  temp=np.sum(np.square(y-ypred))
  cost=(1/n)*temp
  md=-(2/n)*sum((y-ypred)*x)
  bd=-(2/n)*sum(y-ypred)
  m_in=m_in-(learning_rate*md)
  bcur=bcur-(learning_rate*bd)
  print("m {},b {}, cost {},iterations {}".format(m_in,bcur,cost,i))





```

```python
#k means for 2D arrays
import numpy as np
import matplotlib.pyplot as plt
data=np.array([[2,3],[5,6],[8,7],[1,4],[2,2],[6,7],[3,4],[8,6]])
m1=data[0]
m2=data[3]

while True:
  cl_1=[]
  cl_2=[]
  diff1=data-m1
  diff2=data-m2
  dist1=np.sqrt(np.sum(diff1**2,axis=1))
  dist2=np.sqrt(np.sum(diff2**2,axis=1))
  for i in range(len(data)):
    if dist1[i]<dist2[i]:
      cl_1.append(data[i])
    else:
      cl_2.append(data[i])
  cl_1=np.array(cl_1)
  print("one cluster=",cl_1)
  print("end")
  cl_2=np.array(cl_2)
  print("another cluster",cl_2)
  print("end")
  c1_mean=np.mean(cl_1,axis=0)
  print("c1 mean",c1_mean)
  c2_mean=np.mean(cl_2,axis=0)
  print("c2 mean",c2_mean)
  print("iteration completed")
  if np.array_equal(c1_mean, m1) and np.array_equal(c2_mean, m2):

    print("done")
    break
  else:
    m1=c1_mean
    m2=c2_mean
```

```python
#knn algorithm for normal or underweight category
import numpy as np
h=np.array([[167,51],[182,62],[176,69],[173,64],[172,65],[174,56],[169,58],[173,57],[170,55]]) #input
c=([0,1,1,1,1,0,1,1,1])
new_data=[170,57]   #new data point
dist=np.sqrt(np.sum((new_data-h)**2,axis=1))
print(dist)  #calculating distance
index=np.argsort(dist)
print(index)
n=0
u=0
for i in range(6):
  x=index[i]
  if c[x]==1:
    n=n+1
  else:
    u=u+1
if n>u:
  print("the given set belongs to Normal" )
else:
  print("the given set belongs to underweight")
```

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Dense, Flatten
from keras.models import Sequential
from keras.utils import to_categorical
from keras.datasets import mnist
```

```python
# Load MNIST handwritten digit data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# Convert y_train into one-hot format
temp = []
for i in range(len(y_train)):
    temp.append(to_categorical(y_train[i], num_classes=10))

y_train = np.array(temp)
```

```python
# Convert y_test into one-hot format
temp = []
for i in range(len(y_test)):
    temp.append(to_categorical(y_test[i], num_classes=10))

y_test = np.array(temp)
```

```python
# Create simple Neural Network model
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(512, activation='sigmoid'))
model.add(Dense(256, activation='sigmoid'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
	      optimizer='adam',
	      metrics=['acc'])
```

```python
# Train the Neural Network model
train = model.fit(X_train, y_train, epochs=5, validation_data=(X_test,y_test))
```

```python
# Making predictions using our trained model
predictions = model.predict(X_test)
predictions = np.argmax(predictions, axis=1)
```

```python
# Display some predictions on test data
fig, axes = plt.subplots(ncols=10, sharex=False,
			 sharey=True, figsize=(20, 4))
for i in range(10):
	axes[i].set_title(predictions[i])
	axes[i].imshow(X_test[i], cmap='gray')
	axes[i].get_xaxis().set_visible(False)
	axes[i].get_yaxis().set_visible(False)
plt.show()
```

```python
score=model.evaluate(X_test, y_test, verbose=0)
print('Test loss:',score[0])
print('Test accuracy:',score[1])
```
